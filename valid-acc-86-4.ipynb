```python
import tensorflow as tf
import tensorflow_hub as hub
```

```python
train_datagen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1/255)

valid_datagen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1/255)

train_data = train_datagen.flow_from_directory('/kaggle/input/covid19-image-dataset/Covid19-dataset/train',
                                              target_size=(300,300),
                                              class_mode='categorical',
                                              batch_size=32)

valid_data = valid_datagen.flow_from_directory('/kaggle/input/covid19-image-dataset/Covid19-dataset/test',
                                              target_size=(300,300),
                                              class_mode='categorical',
                                              batch_size=32)
```

```python
model = tf.keras.models.Sequential([
    hub.KerasLayer('https://tfhub.dev/google/efficientnet/b1/classification/1',input_shape=(300,300,3),trainable=False),
    tf.keras.layers.Dense(3,activation='softmax')
])

model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(train_data,steps_per_epoch=len(train_data),epochs=10,
                   validation_data=(valid_data),validation_steps=len(valid_data))
```
